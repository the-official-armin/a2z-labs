## Uploading a tokenizer

To upload a tokenizer, it is very similar to uploading a model. We use the ```.from_pretrained``` method to upload the tokenizer automatically from the model we are using.

````python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained('bert-base-cased')

encoded_input = tokenizer('Hello, my name is AK')
print(encoded_input)
````

## Special Tokens

As talked about in one of my learning notes the tokenizer automatically puts in specail tokens to decide where is the start and end of a sentence.

````python
tokenizer.decode(encoded_input['input_ids'])
````

## Encoding a sentence

We can tokenize a sentence independently from the model and can use 2 techinques to format the sentence for the model to easily process the input data: **Padding and truncation**

````python
encoded_input = tokenizer('How are you', 'I am fine, thanks for asking', return_tensors='pt')
print(encoded_input)
````

````python
encoded_input = tokenizer(['How are you?', 'I am fine, thank you'], padding=True, return_tensors='pt')

print(encoded_input)
````

````python
encoded_input = tokenizer(
    ['How are you?', 'I am fine, thanks for asking'],
    padding=True,
    truncation=True,
    max_length = 5,
    return_tensors='pt',
)

print(encoded_input)
````

````python
encoded_input = tokenizer("How are you?")
print(encoded_input["input_ids"])
tokenizer.decode(encoded_input["input_ids"])
````
